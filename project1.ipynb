```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from imblearn.pipeline import Pipeline as ImbPipeline
```

```python
# Define the target variable
df['Match_Outcome'] = df.apply(
    lambda row: 'Home Win' if row['Goals_Home'] > row['Goals_Away']
                else ('Draw' if row['Goals_Home'] == row['Goals_Away']
                      else 'Away Win'), axis=1)
```

```python
# Features and target
features = [
    'Possession_Home', 'Possession_Away', 'Shots_Home', 'Shots_Away', 'Shots_On_Home', 'Shots_On_Away',
    'Shots_Off_Home', 'Shots_Off_Away', 'Shots_Blocked_Home', 'Shots_Blocked_Away', 'Passes_Home', 'Passes_Away',
    'Chances_Home', 'Chances_Away', 'Corners_Home', 'Corners_Away', 'Offside_Home', 'Offside_Away',
    'Tackles_Home', 'Tackles_Away', 'Duels_Home', 'Duels_Away', 'Saves_Home', 'Saves_Away',
    'Fouls_Home', 'Fouls_Away', 'Yellow_Cards_Home', 'Yellow_Cards_Away', 'Red_Cards_Home', 'Red_Cards_Away']
```

```python
X = df[features]
y = df['Match_Outcome']
```

```python
# Encoding the target variable
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)
```

```python
# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Define scaling and models
scaler = StandardScaler()
pipeline_rf = ImbPipeline([
    ('scaler', scaler),
    ('rf', RandomForestClassifier(random_state=42))
])
pipeline_gb = ImbPipeline([
    ('scaler', scaler),
    ('gb', GradientBoostingClassifier(random_state=42))
])
```

```python
# Hyperparameter tuning with GridSearchCV
param_grid_rf = {
    'rf__n_estimators': [100, 200],
    'rf__max_depth': [None, 10, 20]
}
param_grid_gb = {
    'gb__n_estimators': [100, 200],
    'gb__learning_rate': [0.01, 0.1],
    'gb__max_depth': [3, 5]
}
```

```python
grid_search_rf = GridSearchCV(pipeline_rf, param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_gb = GridSearchCV(pipeline_gb, param_grid_gb, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
# Fitting Random Forest
grid_search_rf.fit(X_train_smote, y_train_smote)
```

```python
# Fitting Gradient Boosting
grid_search_gb.fit(X_train_smote, y_train_smote)
```

```python
# Making predictions
rf_predictions = grid_search_rf.predict(X_test)
gb_predictions = grid_search_gb.predict(X_test)
```

```python
# Evaluate the models
def evaluate_model(predictions, y_true, model_name):
    print(f"\n{model_name} Evaluation:")
    print(f"Accuracy: {accuracy_score(y_true, predictions):.2f}")
    print(f"Precision: {precision_score(y_true, predictions, average='weighted'):.2f}")
    print(f"Recall: {recall_score(y_true, predictions, average='weighted'):.2f}")
    print(f"F1 Score: {f1_score(y_true, predictions, average='weighted'):.2f}")
    print("\nClassification Report:")
    print(classification_report(y_true, predictions, target_names=le.classes_))
```

```python
evaluate_model(rf_predictions, y_test, 'Random Forest')
evaluate_model(gb_predictions, y_test, 'Gradient Boosting')
```

```python
# Define the pipeline for SVM
pipeline_svm = ImbPipeline([
    ('scaler', scaler),
    ('svm', SVC(random_state=42))
])
```

```python
# Define the parameter grid for SVM
param_grid_svm = {
    'svm__C': [0.1, 1, 10],
    'svm__kernel': ['linear', 'rbf']
}
```

```python
# Hyperparameter tuning with GridSearchCV
grid_search_svm = GridSearchCV(pipeline_svm, param_grid_svm, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
# Fitting SVM
grid_search_svm.fit(X_train_smote, y_train_smote)
```

```python
# Making predictions with SVM
svm_predictions = grid_search_svm.predict(X_test)
```

```python
# Evaluate the SVM model
evaluate_model(svm_predictions, y_test, 'Support Vector Machine')
```
